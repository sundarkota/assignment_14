```python
import pandas as pd
import numpy as np
import mysql.connector
from sqlalchemy import create_engine
import MySQLdb
```

<h3>1. Create an sqlalchemy engine using a sample from the data set</h3>

```python
conn = create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"
                       .format(user="root",
                               pw="",
                               db="sqldb"))
columns = ["age","workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country","income"]
datset=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", names=columns)
df=pd.DataFrame(datset)
```

```python
df.to_sql(name="adult", con=conn, if_exists = 'replace', index=False)
```

```python
Rdata = pd.read_sql('SELECT * FROM adult limit 10', conn)
Rdata
```
